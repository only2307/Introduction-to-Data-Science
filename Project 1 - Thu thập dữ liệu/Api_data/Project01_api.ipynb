### Get Playlists

```python
# Viết tên các cột csv, yêu cầu cần có thư mục Api_data
with open('./Api_data/playlist.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["artwork_url","created_at","description","duration","embeddable_by","genre","id","kind","label_name","last_modified","license","likes_count","managed_by_feeds","permalink","permalink_url","public","purchase_title","purchase_url","release_date","reposts_count","secret_token","sharing","tag_list","title","uri","user_id","set_type","is_album","published_at","display_date","user","tracks","track_count"])
```

```python
count=1 # Dòng đầu tiên đã viết tên cột
while count<1500:
  response = requests.get('https://api-v2.soundcloud.com/playlists/'+str(x), params=params)
  if x%100==0: # theo dõi và ghi tiến độ
    print("query "+str(x)+" times, get "+str(count)+" playlists")
  # nếu response trả về hợp lệ
  if response.status_code==200:
    data=json.loads(response.content)
    # nếu data hợp lệ
    if data:
      # nếu có tracks
      if data["tracks"]: # tracks chứa thông tin của các track có trong playlist nhưng ở đây chỉ lưu id ra thôi, chính vì vậy tạo ra 1 mảng để chứa
        tracksID=[]
        # tách tracks id vào mảng tracksId
        for e in data["tracks"]:
          tracksID.append(e["id"])
        # biến đổi mảng tracksID thành chuỗi để ghi vào file
        data["tracks"]='-'.join(str(e) for e in tracksID) # ví dụ: [1,3,5] -> '1-3-5' 
        # row là dữ liệu trước khi ghi vô file
        row=[data["artwork_url"],data["created_at"],data["description"],data["duration"],data["embeddable_by"],data["genre"],data["id"],data["kind"],data["label_name"],data["last_modified"],data["license"],data["likes_count"],data["managed_by_feeds"],data["permalink"],data["permalink_url"],data["public"],data["purchase_title"],data["purchase_url"],data["release_date"],data["reposts_count"],data["secret_token"],data["sharing"],data["tag_list"],data["title"],data["uri"],data["user_id"],data["set_type"],data["is_album"],data["published_at"],data["display_date"],data["user"],data["tracks"],data["track_count"]]
        # mở file ra và ghi vô
        with open('./Api_data/playlist.csv', 'a', newline='') as file:
          writer = csv.writer(file)
          writer.writerow(row)
          count=count+1 # tăng biến đếm lên 1
  x=x+1 # tăng id playlist lên 1
```

### Get Tracks

```python
# Đọc file playlist.csv 
df = pd.read_csv('./Api_data/playlist.csv')

# Khởi tạo mảng chứa id của tracks ở trong file csv đó
arrTrackIds=[] # khoảng 8000 id riêng biệt, tuy nhiên ở đây chỉ lấy 1500
# đọc các id từ cột tracks
for e in df['tracks']:# duyệt các phần tử trong cột tracks, e có dạng 241966-241967-241968-241969-241970-618116, tách nhau ra = dấu - thì đây là id của tracks
  trackIds=e.split('-')# trackIds=[241966,241967,241968,241969,241970,618116]
  for i in trackIds:
    if i not in arrTrackIds: # vì các playlist có thể có các tracks giống nhau
      arrTrackIds.append(i) # ví dụ: i = 241966 không nằm trong mảng arrTrackIds thì thêm vào mảng arrTrackIds
```

```python
# viết tên các cột csv, yêu cầu cần có thư mục Api_data 
with open('./Api_data/track.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["artwork_url","caption","commentable","comment_count","created_at","description","downloadable","download_count","duration","full_duration","embeddable_by","genre","has_downloads_left","id","kind","label_name","last_modified","license","likes_count","permalink","permalink_url","playback_count","public","publisher_metadata","purchase_title","purchase_url","release_date","reposts_count","secret_token","sharing","state","streamable","tag_list","title","track_format","uri","urn","user_id","visuals","waveform_url","display_date","media","station_urn","station_permalink","track_authorization","monetization_model","policy","user"])
```

```python
count=1
x=0
while count<1500: #ở đây chỉ cần 1500 bản ghi tracks
  response = requests.get('https://api-v2.soundcloud.com/tracks/'+arrTrackIds[x], params=params)
  if x%100==0: #theo dõi
    print('query '+str(x)+' times, get '+str(count)+' tracks')
  if response.status_code==200:
    data=json.loads(response.content)
    # nếu data hợp lệ
    if data:
      # row là dữ liệu trước khi ghi vô file
        row=[data["artwork_url"],data["caption"],data["commentable"],data["comment_count"],data["created_at"],data["description"],data["downloadable"],data["download_count"],data["duration"],data["full_duration"],data["embeddable_by"],data["genre"],data["has_downloads_left"],data["id"],data["kind"],data["label_name"],data["last_modified"],data["license"],data["likes_count"],data["permalink"],data["permalink_url"],data["playback_count"],data["public"],data["publisher_metadata"],data["purchase_title"],data["purchase_url"],data["release_date"],data["reposts_count"],data["secret_token"],data["sharing"],data["state"],data["streamable"],data["tag_list"],data["title"],data["track_format"],data["uri"],data["urn"],data["user_id"],data["visuals"],data["waveform_url"],data["display_date"],data["media"],data["station_urn"],data["station_permalink"],data["track_authorization"],data["monetization_model"],data["policy"],data["user"]]
        # mở file ra và ghi vô
        with open('./Api_data/track.csv', 'a', newline='') as file:
          writer = csv.writer(file)
          writer.writerow(row)
          count=count+1 # tăng biến đếm lên 1
  x=x+1
```
